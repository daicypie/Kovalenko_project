## Контрольні запитання 
1. У чому відмінність інтерпретатора від компілятора? Python – компілятор чи інтерпретатор?
   
Інтерпретатор працює з кодом поетапно, виконуючи його рядок за рядком, на відміну від компілятора, який перекладає весь код одночасно. Python використовує інтерпретатор для виконання своїх програм.

2. Що таке IDE PyCharm?
 
PyCharm – це інтегроване середовище розробки (IDE), спеціально створене для програмування на Python. Воно пітримує розробку, як наукових так і веб проєктів.

3. Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ SVN та GIT.

Система контролю версій – це інструмент, який дозволяє зберігати різні версії файлу, ніби ми створюємо фотографії його стану в різний час. Це дозволяє повернутися до попередніх версій, порівняти зміни та відстежувати історію розробки.

SVN — це централізована система контролю версій, у якому всі дані зберігаються на одному сервері 

Git — це розподілена система контролю версій, де кожен користувач має повну копію репозиторію

4. Що відбувається при виконанні команди `git add.` ?

Додавання всіх файлів в `commit` 

5. Що відбувається при виконанні команди `git commit`?

Створюється знімок змін у файлах, які додані до індексу, і зберігаються в локальному репозиторії

6. Якою Git-командою виконується розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?

Командою `Push`
